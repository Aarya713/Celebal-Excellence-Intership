### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

A matrix geometrically represents a transformation of space, and when it “stretches a vector,” it means the matrix is changing the vector’s length while preserving or altering its direction.

Think of a vector as an arrow drawn from the origin. When you multiply it by a matrix, the space itself gets transformed:

In some directions, the matrix pulls points away from the origin, making vectors longer (this is stretching).
In other directions, it may push points closer, making vectors shorter (compression).
The direction of the vector may also change depending on how the matrix rotates or skews space.

So geometrically, “stretching” means the matrix is acting like a rubber sheet transformation, where the sheet is pulled more in certain directions than others.

Some special directions (eigenvectors) behave simply: they don’t change direction at all—they only get scaled up or down by a numbe



````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = None, None  # YOUR CODE HERE

eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]

    Av = A @ v
    lv = lam * v

    print(f"\nEigenpair {i+1}:")
    print("A @ v =", Av)
    print("λ * v =", lv)
    assert np.allclose(Av, lv)

# Plot
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]

    ax.quiver(
        0, 0,
        lam * v[0],
        lam * v[1],
        angles='xy',
        scale_units='xy',
        scale=1,
        label=f'λ={lam:.2f}'
    )

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Eigenvectors Scaled by Eigenvalues')
ax.axhline(0, color='k', lw=0.8)
ax.axvline(0, color='k', lw=0.8)
ax.legend()
plt.grid(True)
plt.show()

print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

When you multiply a matrix by a vector, the vector usually changes both direction and length. But there are special vectors called eigenvectors that do something unique: they do not change direction at all after the transformation. They only get stretched or squished.

The amount of stretching or squishing is given by the eigenvalue.

So geometrically:

A matrix acts like a transformation that reshapes space (like stretching a rubber sheet).
Most vectors get rotated or bent in new directions.
But eigenvectors are “special directions” in that space that stay on the same line.
Along those directions, the matrix only scales them:
Eigenvalue > 1 → stretching (vector becomes longer)
0 < Eigenvalue < 1 → squishing (vector becomes shorter)
Eigenvalue < 0 → flipping + scaling

A one-sample t-test compares the mean salary of Engineering employees with the overall company mean salary to check whether the observed difference is statistically significant or just due to random variation.

### 5.5 Model Monitoring Concepts

1. Implement `compute_psi(expected, actual, bins=10)` that computes the **Population Stability Index** between two distributions.
2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

Concept Drift

Concept drift occurs when the relationship between input features (X) and the target variable (Y) changes over time, meaning the model’s learned patterns become outdated even if the input distribution stays similar.

Covariate Drift

Covariate drift happens when the distribution of input features (P(X)) changes over time, but the relationship between inputs and output (P(Y|X)) remains the same. In this case, the model may perform poorly because it is seeing different kinds of data than it was trained on.

PSI (Population Stability Index) Thresholds

PSI measures how much a feature’s distribution has shifted over time.

PSI < 0.10 → No significant shift (stable population)
0.10 ≤ PSI < 0.25 → Moderate / minor shift (monitor closely)
PSI ≥ 0.25 → Major shift (significant distribution change, model risk)

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    # YOUR CODE HERE

    epsilon = 1e-10

    breakpoints = np.linspace(np.min(expected), np.max(expected), bins + 1)

    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _ = np.histogram(actual, bins=breakpoints)

    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    expected_pct = expected_pct + epsilon
    actual_pct = actual_pct + epsilon

    psi = np.sum((actual_pct - expected_pct) *
                 np.log(actual_pct / expected_pct))

    return psi

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
# YOUR CODE HERE

ax.hist(train_dist, bins=20, alpha=0.5, density=True, label='Training Distribution')
ax.hist(drift_dist, bins=20, alpha=0.5, density=True, label='Production Distribution')

ax.legend()
ax.set_title('Distribution Shift Visualization')
plt.tight_layout()
plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**



Concept drift: Concept drift occurs when the relationship between input features (X) and the target variable (Y) changes over time, meaning the model’s learned patterns become outdated even if the input distribution stays similar.

Covariate drift: Covariate drift happens when the distribution of input features P(X) changes over time while the relationship P(Y|X) remains the same, causing the model to see data that differs from what it was trained on.

PSI Thresholds

PSI < 0.1: No significant shift; the feature distribution is stable and the model is reliable.

PSI 0.1–0.2: Moderate or minor shift; changes are noticeable, so the feature should be monitored closely for potential impact.

PSI > 0.2: Significant shift; strong distribution change indicating possible model degradation risk.

Retraining Trigger Example

A model should be retrained when PSI exceeds 0.2 for important features or when model performance metrics (like accuracy, F1-score, or AUC) start degrading significantly over time. For example, if customer income distribution shifts (PSI > 0.2) and simultaneously loan default prediction accuracy drops, it indicates that the model is no longer aligned with current data patterns and must be retrained.

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

Bayesian Terms Mapped to Spam Detection Example

In a spam filter, we classify an email as Spam or Not Spam using Bayes’ Theorem:

Prior

The prior is the initial probability before seeing any email content.
Example:

P(Spam) = probability that any random email is spam
P(Not Spam) = probability that any email is legitimate

 This comes from historical data (e.g., 40% of emails are spam).

Likelihood

The likelihood is the probability of observing the email content given the class.
Example:

P(words like "win money", "free", "offer"∣Spam)
P(words like "meeting", "project"∣Not Spam)

 It tells how likely certain words appear in spam vs non-spam emails.

Evidence (Marginal Likelihood)

The evidence is the overall probability of seeing the email content, regardless of class.
Example:

P(email content)

 It acts as a normalization factor and ensures probabilities sum correctly across spam and not spam cases.

Posterior

The posterior is the updated probability after seeing the email content.
Example:

P(Spam∣email content)

 This is the final result used by the model to decide whether the email is spam or not

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = None  # YOUR CODE HERE

p_free = (p_free_given_spam * p_spam) + (p_free_given_ham * p_ham)

# Bayes
p_spam_given_free = None  # YOUR CODE HERE

p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    # YOUR CODE HERE
    p_ham = 1 - prior_spam
    p_word = (p_word_given_spam * prior_spam) + (p_word_given_ham * p_ham)
    posterior = (p_word_given_spam * prior_spam) / p_word
    return posterior

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means = None  # YOUR CODE HERE — array of 5000 sample means

sample_means = np.array([
    np.random.choice(population, size=sample_size).mean()
    for _ in range(n_samples)
])

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = None  # YOUR CODE HERE — predicted std of sample means

clt_std = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()

# YOUR CODE HERE — histogram of sample means + theoretical normal overlay
ax.hist(sample_means, bins=30, density=True, alpha=0.6, label='Sample Means')

x = np.linspace(sample_means.min(), sample_means.max(), 300)
ax.plot(
    x,
    stats.norm.pdf(x, loc=pop_mean, scale=clt_std),
    linewidth=2,
    label='CLT Normal Approximation'
)

ax.legend()
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout()
plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````
